# 11 데이터정렬

```python
import pandas as pd

df = pd.read_csv("score.csv")
df
```

```python
df.sort_values('키', ascending = False)
```

```python
df.sort_values(['수학', '영어']) # 수학을 기준으로, 같으면 영어 순으로
```

```python
df.sort_values(['수학', '영어'], ascending = [True, False]) # 수학점수는 오름차순, 영어점수는 내림차순으로 하고 싶을때
```
