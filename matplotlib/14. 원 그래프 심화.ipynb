```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #windows
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['py', 'java', 'js', 'c', 'c+', 'ETC']
colors = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff', '#a0c4ff']
explode= [.05]*6

plt.pie(values, labels=labels, autopct='%.1f%%', colors=colors, explode=explode)
plt.show()
```

## 그래프 스타일

```python
wedgeprops={'width':.8, 'edgecolor':'w', 'linewidth':2}
plt.pie(values, labels=labels, autopct='%.1f%%', colors=colors, wedgeprops=wedgeprops)
plt.show()
```

```python
def custom_autopct(pct):
    return ('%.1f%%' % pct) if pct >= 10 else ''

plt.pie(values, labels=labels, autopct=custom_autopct, colors=colors, wedgeprops=wedgeprops)
plt.show()
```
