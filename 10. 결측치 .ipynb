## 데이터 제외하기 dropna

```python
import pandas as pd

df = pd.read_csv("score.csv")
df
```

```python
df.dropna() ## nan 포함하는 데이터 삭제
```

- axis : index or column   (행을 지울지, 열을 지울지)
- how : any or all

```python
df.dropna(axis = 'index', how = 'any') 
```

```python
df.dropna(axis = 'columns', how = 'any') # nan를 포함하고 있는 열 제외, 하나라도 포함하고 있으면
```

```python
df['학교'] = np.nan
```

```python
df.dropna(axis = 'columns', how = 'all') # 컬럼 전체가 nan일때 해당열 삭제
```
