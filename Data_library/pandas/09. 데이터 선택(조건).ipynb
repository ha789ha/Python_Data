# 9. 데이터선택(조건)

```python
import pandas as pd

df = pd.read_csv("score.csv")
df
```

```python
filt = (df['키'] >= 185) # 185이상인 행들 선택
df[filt]
```

```python
df[~filt] # ~를 쓰면 이것을 제외하고 보여준다
```

```python
df.loc[df['키'] >= 185, ['이름', '수학']] # 이런 식으로 응용 가능
```

## 다양한 조건

### and

```python
df.loc[(df['키'] >= 185) & (df['학교'] == '북산고')]
```

### OR

```python
df.loc[(df['키'] >= 185) | (df['학교'] == '북산고')]
```

## str 함수

```python
filt = df['이름'].str.startswith('송') # 송씨 성을 가진 사람
df[filt]
```

```python
filt = df['이름'].str.contains('태')
df[filt]
```

```python
langs = ['Python', 'Java']
filt = df['SW특기'].isin(langs) # 특기가 파이썬과 자바인 사람 선택, 황태산은 python이 모두 대문자라 뽑히지 않음
df[filt]
```

```python
langs = ['Python', 'Java']
filt = df['SW특기'].str.lower().isin(langs) # 이렇게 하면 대문자까지 모두 뽑을수 있다
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java', na = True) # NAN 데이터에 대해 모두 true로 변환
df[filt]
```
