# 12. 데이터수정

```python
import pandas as pd

df = pd.read_csv("score.csv")
df.set_index('지원번호')
```

## column 수정

```python
df['학교'].replace({'북산고': '상북고', '능남고' : '반송고'}) ## inplace 잊지말것!
```

```python
df['학교'].replace({'북산고': '상북고'}, inplace = True)
```

```python
df['SW특기'] = df["SW특기"].str.lower() # 전체 데이터를 바꿀 수도 있다
df
```

```python
df['학교'] = df['학교'] + '등학교' # 학교에 문자열 추가해주고 싶을때
df
```

## column 추가


```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
df['결과'] = 'fail'
df.loc[df['총합'] >= 400, '결과'] = 'pass'  # 총합이 400 넘는 사람들을 패스로 해놓는 것
df
```

## column 삭제

```python
df.drop(columns = ['총합']) # 총합 columns을 삭제
```

## ROw 삭제

```python
df.drop(index = '4번')
```

```python
filt = df['수학'] < 80
df[filt].index
df.drop(index = df[filt].index )
```

## Row 추가

```python
df.loc['9번'] = ['이정환', '해남고', 184, 90, 90, 90, 90, 'kotlin', 450, 'pass']
df
```

## cell 수정

```python
df.loc['4번', 'SW특기'] = 'Python'
```

```python
df.loc['5번', ['학교', 'SW특기']] = ['능남고', 'C']
```
