# 04. 스타일

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #windows
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False

x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y, linewidth=2, marker ='o') # 라인 두께 조절과 marker
plt.show()
```

## Marker

```python
plt.plot(x, y, marker='o', markersize=10, linestyle='none') # 마커만 남김 
```

```python
plt.plot(x, y, marker='o', markersize=10, markeredgecolor='red', markerfacecolor='yellow') # 마커의 스타일 조정
```

## 선 스타일

```python
plt.plot(x, y, linestyle= '--')
```

## 색깔

```python
plt.plot(x, y, color='pink')
```

## format

```python
plt.plot(x, y, 'ro--') # color, marker, linestyle 한번에 지정 가능
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x, y, 'go')
```

## 축약어

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, mec='blue', ls=':')
```

## 투명도

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, mec='blue', ls=':', alpha=.2)
```

## 그래프 크기

```python
plt.figure(figsize=(10, 5), dpi=200) # 해상도 조절
plt.plot(x, y)
```

## 배경색

```python
plt.figure(facecolor='red')
plt.plot(x, y)
```
