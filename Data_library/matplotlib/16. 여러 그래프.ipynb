# 16. 여러 그래프

```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #windows
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('data.csv')
```

```python
fig, axs = plt.subplots(2, 2, figsize= (15, 10)) ## 2*2의 그래프를 그림
fig.suptitle('여러 그래프 넣기')

# 첫번째 그래프;
axs[0, 0].bar(df['이름'], df['국어'], label='국어 점수')
axs[0, 0].set_title('첫번째 그래프')
axs[0, 0].legend()
axs[0, 0].set(xlabel='이름', ylabel='점수')
axs[0, 0].set_facecolor('lightyellow')
axs[0, 0].grid(linestyle='--', linewidth=.5)

# 두 번째 그래프
axs[0, 1].plot(df['이름'], df['수학'], label='수학')
axs[0, 1].plot(df['이름'], df['영어'], label='영어')
axs[0, 1].legend()

# 세 번째 그래프
axs[1, 0].barh(df['이름'], df['키'])

# 네 번째 그래프
axs[1, 1].plot(df['이름'], df['사회'], color='g', alpha=.4)
```
