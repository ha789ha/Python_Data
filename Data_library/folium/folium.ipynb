# folium 라이브러리

```python
import folium as fl
```

### 기본 좌표 설정

```python
latitude = 35.219086
longitude = 128.9510819
```

```python
m = fl.Map(
    location=[latitude, longitude],
    zoom_start=17,
    width=1000,
    height=700
)
m
```

### 마커 추가

```python
fl.Marker([latitude, longitude],
popup="오성면", # 표기할 팝업 문구
tooltip="오성면 근처",
icon=fl.Icon('red', icon='star')).add_to(m) # 마우스 오버시 표기되는 문구
m
```

```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
df.head(5)
```

```python
df.columns
df.행정동명
```

```python
sub = df.loc[df["행정동명"].isin(['삼안동', "청도면", "삼안동"])]
```

```python
sub[['위도', '경도', '상호명']]
```

```python
from folium.plugins import MarkerCluster

m = fl.Map(
    location=[latitude, longitude],
    zoom_start=15
)

coords = sub[['위도', '경도']]

marker_cluster = MarkerCluster().add_to(m)

for lat, long in zip(coords['위도'], coords['경도']):
    fl.Marker([lat, long], icon=fl.Icon(color="green")).add_to(marker_cluster)

m
```
